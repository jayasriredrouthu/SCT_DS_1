```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load Dataset
df = pd.read_csv("indian_developer_burnout_2026.csv")

# Display first 5 rows
print(df.head())

# Check dataset information
print(df.info())
```

```python
# Remove missing values from Age, Gender and Burnout Category

df = df.dropna(subset=["age", "gender", "burnout_risk_category"])

# Check shape after cleaning
print("Dataset Shape:", df.shape)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["age"],
    bins=10,
    edgecolor="black"
)

plt.title("Age Distribution of Indian Developers")
plt.xlabel("Age")
plt.ylabel("Number of Developers")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python


# --------------------------
# Gender Distribution
# --------------------------

gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6,5))

colors = ["blue", "orange", "green", "red", "purple"]

bars = plt.bar(
    gender_counts.index,
    gender_counts.values,
    color=colors[:len(gender_counts)],
    edgecolor="black",
    linewidth=1.5
)

# Value labels on bars
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        str(int(height)),
        ha="center"
    )

plt.title("Gender Distribution of Indian Developers")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Bar Chart: Burnout Risk Category Distribution
# -------------------------------

burnout_counts = df["burnout_risk_category"].value_counts()

plt.figure(figsize=(7,5))

plt.bar(
    burnout_counts.index,
    burnout_counts.values,
    edgecolor="black"
)

# Display values on bars
for i, value in enumerate(burnout_counts.values):
    plt.text(i, value + 1, str(value), ha="center", fontsize=10)

plt.title("Burnout Risk Category Distribution", fontsize=15)
plt.xlabel("Burnout Risk Category")
plt.ylabel("Number of Developers")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
